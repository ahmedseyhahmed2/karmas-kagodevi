```python
!pip install networkx

import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community

%matplotlib inline
```

```python
nodes_df = pd.read_csv("InputFileNodes.csv")
edges_df = pd.read_csv("InputFileEdges.csv")

print("Düğümler:")
display(nodes_df.head())

print("Kenarlar:")
display(edges_df.head())
```

```python
# forward edges: from-to
edges_clean = edges_df[['from', 'to']].copy()

# yönsüz için duplicate kaldırma
edges_undirected = edges_clean.apply(
    lambda row: tuple(sorted([row['from'], row['to']])), axis=1
)
edges_undirected = list(set(edges_undirected))

print("Temizlenmiş yönsüz kenar sayısı:", len(edges_undirected))
```

```python
# Yönsüz graph G
G = nx.Graph()
G.add_nodes_from(nodes_df['id'])
G.add_edges_from(edges_undirected)

# Yönlü graph G_directed
G_directed = nx.DiGraph()
G_directed.add_nodes_from(nodes_df['id'])
G_directed.add_edges_from(edges_df[['from','to']].itertuples(index=False, name=None))

print("G -> düğüm:", G.number_of_nodes(), " kenar:", G.number_of_edges())
print("G_directed -> düğüm:", G_directed.number_of_nodes(), " kenar:", G_directed.number_of_edges())
```

```python
plt.figure(figsize=(8,6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, node_size=500, with_labels=True)
plt.title("Yönsüz Graph G")
plt.show()
```

```python
plt.figure(figsize=(8,6))
pos2 = nx.spring_layout(G_directed, seed=42)
nx.draw(G_directed, pos2, node_size=500, with_labels=True, arrows=True)
plt.title("Yönlü Graph G_directed")
plt.show()
```

```python
# Dereceler
degrees = dict(G.degree())
degree_df = pd.DataFrame({"Node": degrees.keys(), "Degree": degrees.values()})
display(degree_df)

# Ortalama derece
avg_degree = sum(degrees.values()) / G.number_of_nodes()
print("Ortalama Derece:", avg_degree)
```

```python
plt.figure(figsize=(7,5))
plt.hist(list(degrees.values()), bins=10, rwidth=0.8)
plt.xlabel("Derece")
plt.ylabel("Frekans")
plt.title("Derece Histogramı (G)")
plt.grid(axis="y", alpha=0.3)
plt.show()
```

```python
# 9 - Degree centrality
deg_centrality = nx.degree_centrality(G)

# 10 - Closeness centrality
clo_centrality = nx.closeness_centrality(G)

# 11 - Betweenness centrality
bet_centrality = nx.betweenness_centrality(G)

# 12 - Eigenvector centrality
eig_centrality = nx.eigenvector_centrality(G)

# 13 - Katz centrality
katz_centrality = nx.katz_centrality(G, alpha=0.1, beta=1.0)

# 14 - PageRank
pagerank = nx.pagerank(G)

centrality_table = pd.DataFrame({
    "Node": list(G.nodes()),
    "Degree": [degrees[n] for n in G.nodes()],
    "DegreeCentrality": [deg_centrality[n] for n in G.nodes()],
    "ClosenessCentrality": [clo_centrality[n] for n in G.nodes()],
    "BetweennessCentrality": [bet_centrality[n] for n in G.nodes()],
    "EigenvectorCentrality": [eig_centrality[n] for n in G.nodes()],
    "KatzCentrality": [katz_centrality[n] for n in G.nodes()],
    "PageRank": [pagerank[n] for n in G.nodes()]
})

display(centrality_table)
```

```python
# 15 - Modularity
communities = community.greedy_modularity_communities(G)
modularity_value = community.modularity(G, communities)
print("Modularity:", modularity_value)

# 16 - Density
density_value = nx.density(G)
print("Density:", density_value)

# 17 - Average clustering
avg_clustering = nx.average_clustering(G)
print("Average Clustering:", avg_clustering)

# 18 - Diameter
if nx.is_connected(G):
    diameter_value = nx.diameter(G)
else:
    # bağlı komponent değilse en büyük component alınır
    largest_cc = max(nx.connected_components(G), key=len)
    diameter_value = nx.diameter(G.subgraph(largest_cc))
    print("UYARI: G bağlı değil; en büyük bileşenin diameter'ı alındı.")

print("Diameter:", diameter_value)
```

```python
degree_values = list(degrees.values())
max_degree = max(degree_values)
deg_count = {k: degree_values.count(k) for k in range(max_degree+1)}
deg_prob = {k: v / G.number_of_nodes() for k, v in deg_count.items()}

plt.figure(figsize=(7,5))
plt.bar(deg_prob.keys(), deg_prob.values())
plt.xlabel("Derece")
plt.ylabel("Olasılık")
plt.title("Derece Dağılımı (Probability Distribution)")
plt.show()
```

```python
in_deg = dict(G_directed.in_degree())
out_deg = dict(G_directed.out_degree())

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(list(in_deg.values()), rwidth=0.8)
plt.title("In-Degree Histogram (G_directed)")
plt.xlabel("In-degree")
plt.ylabel("Frequency")

plt.subplot(1,2,2)
plt.hist(list(out_deg.values()), rwidth=0.8)
plt.title("Out-Degree Histogram (G_directed)")
plt.xlabel("Out-degree")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# 21 - En Kritik 5 Köprü Düğümü (Betweenness Centrality)

betweenness = nx.betweenness_centrality(G)

# Büyükten küçüğe sırala ve ilk 5'i al
critical_bridges = sorted(betweenness.items(), key=lambda x: x[1], reverse=True)[:5]

print("🔵 G Network - En Kritik 5 Köprü Düğümü (Betweenness Centrality):")
for node, value in critical_bridges:
    print(f"{node}  -->  Betweenness: {value:.6f}")
```

```python
# 22 - En Kritik 5 'Lider' Düğümü

# Liderlik skorunu hesapla (ortalama veya ağırlıklı)
leader_score = {
    n: (eig_centrality[n] + pagerank[n] + deg_centrality[n]) / 3
    for n in G.nodes()
}

# Büyükten küçüğe sırala, ilk 5 lideri al
top_5_leaders = sorted(leader_score.items(), key=lambda x: x[1], reverse=True)[:5]

print("🔵 G Network - En Kritik 5 LİDER Düğüm:")
for node, score in top_5_leaders:
    print(f"{node}  -->  Liderlik Skoru: {score:.6f}")
```

```python
# 23 - G_directed network için derece dağılımı

# In-degree ve out-degree hesaplama
in_degrees = dict(G_directed.in_degree())
out_degrees = dict(G_directed.out_degree())

plt.figure(figsize=(12, 5))

# In-degree histogram
plt.subplot(1, 2, 1)
plt.hist(list(in_degrees.values()), rwidth=0.8, color='skyblue')
plt.title("G_directed – In-Degree Dağılımı")
plt.xlabel("In-Degree")
plt.ylabel("Frekans")
plt.grid(axis='y', alpha=0.3)

# Out-degree histogram
plt.subplot(1, 2, 2)
plt.hist(list(out_degrees.values()), rwidth=0.8, color='salmon')
plt.title("G_directed – Out-Degree Dağılımı")
plt.xlabel("Out-Degree")
plt.ylabel("Frekans")
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# 24 - G network için en güçlü 3 düğümü çıkarıp yeni ağı çizdirme

# Güç skoru: Degree + Eigenvector + PageRank ortalaması
power_score = {
    n: (degrees[n] + eig_centrality[n] + pagerank[n]) / 3
    for n in G.nodes()
}

# En güçlü 3 düğümü sırala
strongest_3 = sorted(power_score.items(), key=lambda x: x[1], reverse=True)[:3]
nodes_to_remove = [n for n, score in strongest_3]

print("🔵 En Güçlü 3 Düğüm (Kaldırılacak):")
for node, score in strongest_3:
    print(f"{node}  --> Güç Skoru: {score:.6f}")

# Yeni graph oluştur
G_removed = G.copy()
G_removed.remove_nodes_from(nodes_to_remove)

# Yeni ağı çiz
plt.figure(figsize=(8,6))
pos_new = nx.spring_layout(G_removed, seed=42)
nx.draw(G_removed, pos_new, with_labels=True, node_size=500)
plt.title("En Güçlü 3 Düğüm Kaldırıldıktan Sonra G Ağı")
plt.show()
```
